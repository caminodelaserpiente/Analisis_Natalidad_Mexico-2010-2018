```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib.cm as cm
```

```python
data = pd.read_csv("/content/Natalidad_Mexico.csv")
dataframe = pd.DataFrame(data)
dataframe
```

## Natalidad de los estados de la republica con relación a un año epesífico

```python
# Filtro por año
año = 2016
year_filter = dataframe["year"] == año

# Filtrar desendentemente por estados
df = dataframe.loc[year_filter,["estado","Total","Hombres","Mujeres","year"]]
df = df.sort_values("Total", ascending=False)
df
```

```python
estados = df.loc[:,"estado"]
estados = list(estados)

total =df.loc[:,"Total"]
total = list(total)

print(estados,total)
```

```python
# Graphic list
colors = cm.rainbow(np.linspace(1, 0, 33))
rcParams['figure.figsize'] = 13, 8

plt.title('Natalidad ' + str(año))
plt.xlabel('total nacimientos')
plt.ylabel('estados')
plt.barh(estados, total, color=colors)
```

## Natalidad de un estado con relación a los años transcurridos

```python
estado = "San Luis Potosí"
estado_filter = dataframe["estado"] == estado

df = dataframe.loc[estado_filter,["estado","Total","Hombres","Mujeres","year"]]
df = df.sort_values("Total", ascending=False)
df
```

```python
year = df.loc[:,"year"]
year = np.array(year)
year = year.astype(np.string_)

total =df.loc[:,"Total"]
total = list(total)

print(year,total)
```

```python
# Graphic list
colors = cm.rainbow(np.linspace(1, 0, 9))
rcParams['figure.figsize'] = 13, 8

plt.title('Natalidad ' + str(estado))
plt.xlabel('total nacimientos')
plt.ylabel('year')
plt.barh(year, total, color=colors)
```

## Comparación de Natalidad por estado H,M a los años transcurridos

```python
estado = "Chiapas"
estado_filter = dataframe["estado"] == estado

df = dataframe.loc[estado_filter,["estado","Total","Hombres","Mujeres","year"]]
#df = df.sort_values("Total", ascending=False)
df
```

```python
year = df.loc[:,"year"]
year = np.array(year)
year = year.astype(np.string_)

total =df.loc[:,"Total"]
total = list(total)

h = df.loc[:,"Hombres"]
h = list(h)

m =df.loc[:,"Mujeres"]
m = list(m)

print(year,total,h,m)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(year,h, "b")
axes[0].set_title("Natalidad '" + str(estado)+ "' Hombres")
axes[0].set_xlabel("años")
axes[0].set_ylabel("total natalidad")

axes[1].plot(year,m, "#FF0080")
axes[1].set_title("Natalidad '" + str(estado)+ "' Mujeres")
axes[1].set_xlabel("años")
axes[1].set_ylabel("total natalidad")
```
